````python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")
````

````python
fs  = 100             #Frequncia de amostragem
Ts  = 1/fs            #Perodo de amostragem

f0  = 5.5             #Frequncia fundamental da onda
phi = -np.pi/2        #Fase da onda

t = np.arange(0,2,Ts)        #Vetor de tempo
x = np.cos(2*np.pi*f0*t+phi) #Onda cosseno

N = len(t)            #Nmero de amostras  

fig = plt.figure(figsize=(20,4))

plt.plot(t,x)
plt.ylabel("Amplitude")
plt.xlabel("Tempo (s)")
plt.show()
````

````python
#Transformada Discreta de Fourier utilizando fft
print(f"Tamanho do sinal: {N}")
X = np.fft.fft(x)
# um nmero complexo
print(type(X[0]),X[0])
````

<span style="color:Orange"><b>O algoritmo da FFT necessita que $N$ seja uma potncia de 2 para um clculo eficiente da transformada. Pode-se fazer zero padding para que N seja uma potncia de 2.</span>
    
> FFT (Fast Fourier Transform) refers to a way the discrete Fourier Transform (DFT) can be calculated efficiently, by using symmetries in the calculated terms. The symmetry is highest when n is a power of 2, and the transform is therefore most efficient for these sizes.

### **Mdulo do Sinal** 

````python
plt.stem(np.abs(X))
plt.ylabel("Mdulo")
plt.show()
````

A transformada de fourier de um sinal real  uma funo par, isto , simtrico em relao ao eixo $y$. Entretanto, normalmente no apresentamos dessa maneira o mdulo, para corrigir precisamos calcular as frequncias.

````python
#Calcula o indice de frequncias
freqs = np.fft.fftfreq(N,Ts)

plt.stem(freqs,np.abs(X))
plt.ylabel("Mdulo")
plt.show()
````

As vezes queremos mostrar o eixo y em decibels

````python
#Passa para decibel
X_dec = 20*np.log10(np.abs(X))
plt.stem(freqs, X_dec)
plt.ylabel("Mdulo")
plt.show()
````

Fica dessa forma porque no  exatamente zero o valor nas outras componentes (erro numrioco). Agora perceba que a gente s precisa de metade desse espectro, pois existe simetria!

````python
#Passa para decibel
plt.stem(freqs[:int(N/2)], np.abs(X[:int(N/2)]))
plt.ylabel("Mdulo")
plt.show()
````

### **Fase do sinal** 

Recebe uma sequncia e retorna os ngulos em **radianos (Default)** ou **graus**.

````python
print(f"ngulos em Radianos: {np.round(np.angle(x)[5:15],2)}")

print(f"ngulos em Graus: {np.round(np.angle(x,deg=True)[5:15],2)}")
````

>Unwrap **radian phase** by changing absolute jumps greater than ***discont*** to their 2*pi complement

Quando calculamos o ngulo de um nmero complexo estamos fazendo $(tan\frac{a}{b})$, mas se $\frac{a}{b}$ for mltiplo de $2\pi k, k \in \mathbb{Z}$, temos que o resultado  0, isto , a funo no mostra que estamos dando uma volta no crculo trigonomtrico. Esse fato gera descontinuidades no grfico dos ngulos por isso que para entender ```delay``` precisamos fazer o ```unwrap``` do vetor de ngulos.

````python
fig,axs = plt.subplots(1,2,figsize=(20,4))

#S precisamos da metade!
X_ang = np.angle(X[:int(N/2)+1])
axs[0].plot(X_ang)
axs[1].plot(np.unwrap(X_ang))

axs[0].set_title("Sem unwrap")
axs[1].set_title("Com unwrap")

for ax in axs.flatten():
    ax.set_ylabel("ngulo em radianos") 
    ax.set_xlabel("Frequncias")

plt.tight_layout()
plt.show()
````

<span style="color:Orange"><b>Toda vez que a diferena entre x[1] - x[0] for maior que $\pi$ o unwrap ocorre. Como calcular a fase em graus? </span>

````python
fig,axs = plt.subplots(1,2,figsize=(20,4))

#S precisamos da metade!
X_ang = np.angle(X[:int(N/2)],deg=True)
axs[0].plot(X_ang)
axs[1].plot(np.unwrap(X_ang, discont=180))

axs[0].set_title("Sem unwrap")
axs[1].set_title("Com unwrap")

for ax in axs.flatten():
    ax.set_ylabel("ngulo em radianos") 
    ax.set_xlabel("Frequncias")

plt.tight_layout()
plt.show()
````
