```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")
from scipy import signal
```

## **Nyquist Aliasing**

```python
#Sinal original
N  = 10000

f0 = 10
fs = 40

#Períodos de amostragem
Tr = 1/N
Ts = 1/fs

#Tempo inicial e final
Ti = 0
Tf = 1

tr  = np.arange(Ti,Tf, Tr)
xr  = signal.sawtooth(2 * np.pi * f0 * tr)

ts  = np.arange(Ti,Tf, Ts)
xs  = signal.sawtooth(2 * np.pi * f0 * ts)

fig = plt.figure(figsize = (20,4))

plt.plot(tr, xr)
plt.stem(ts, xs)

plt.ylabel("Amplitude")
plt.xlabel("Tempo(s)")
plt.show()
```

```python
#Selecionando 1 período
tny = ts[int(0.4*fs):int(0.5*fs)]
xny = xs[int(0.4*fs):int(0.5*fs)]
print(f"Quantidade de pontos: {len(tny)}, Período: {len(tny)*1/fs}")
```

```python
X = np.fft.fft(xny)
freqs = np.fft.fftfreq(len(xny), 1/fs)
```

```python
plt.stem(freqs,np.abs(X))
plt.show()
```

```python
xifft = np.fft.ifft(X)
```

```python
plt.plot(np.real(xifft))
plt.show()
```
