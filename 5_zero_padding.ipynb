```python
import os
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")

import librosa

import IPython.display as ipd
```

```python
path = os.path.join("data","186942__lemoncreme__piano-melody.wav")

x, fs = librosa.load(path)

ipd.Audio(path) # load a local WAV file
```

```python
N = len(x)
Ti = 0
Tf = N/fs
t = np.arange(Ti,Tf,1/fs)
```

```python
fig = plt.figure(figsize=(20,4))

plt.plot(t,x)

plt.title("Som real",fontsize = 20)
plt.ylabel("Amplitude",fontsize = 18)
plt.xlabel("Tempo (s)",fontsize = 18)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.show()
```

```python
X = np.fft.fft(x)
freqs = np.fft.fftfreq(N,1/fs)
```

```python
fig, axs = plt.subplots(1,2,figsize=(20,4))

axs[0].plot(freqs[:int(N/2)], np.abs(X[:int(N/2)]))
axs[0].set_title("DFT",fontsize = 20)
axs[0].set_ylabel("Amplitude",fontsize = 18)
axs[0].set_xlabel("Frequência (Hz)",fontsize = 18)

axs[1].plot(freqs[:int(N/2)], np.abs(X[:int(N/2)]))
axs[1].set_title("DFT",fontsize = 20)
axs[1].set_ylabel("Amplitude",fontsize = 18)
axs[1].set_xlabel("Frequência (Hz)",fontsize = 18)
axs[1].set_xlim(0,1000)
plt.show()
```
